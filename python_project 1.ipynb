### 1.how many are there in each team and the percentage splitting w.r.t the total employees


```python
team_members=data['Team'].value_counts()
team_members
```

```python
total_employees=len(data)
total_employees
```

```python
percentage=(team_members/total_employees)*100
percentage
```

### 2. segregate the employees w.r.t different positions

```python
position_group=data.groupby("Position")["Name"].apply((list))
for position,names in position_group.items():
    print (f"Employees in positions {position}:")
    for name in names:
        print(name)
        
```

### 3. find from which age group most of the employees belongs to?

```python
age_group=data["Age"].value_counts()
age_group
```

```python
common_age_group=age_group.idxmax()
print("age group in which most of the belongs to is  :" ,common_age_group )
```

### 4. find out under which team and position, spending interm of salary is high?

```python
high_salary=data.groupby(['Team','Position'])['Salary'].sum()
high_salary.idxmax()
```

### 5. find if there is any correlation between age and salary,represent it visually

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
correlation=data['Age'].corr(data['Salary'])
print(f'correlation between age and salary= ',correlation)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data['Age'],data['Salary'],color='red')
plt.title('correlation between age and salary')
plt.xlabel('age')
plt.ylabel('salary')
```
